Since there are too many unique cars, compress into just the car brand.

```python
df["CarName"].unique()
```

```python
def strip_car_name(car):
    #split on whitespace and get first value (main brand)
    return car.split(" ")[0].lower()

df["CarName"] = df["CarName"].apply(strip_car_name)
```

```python
df["CarName"].unique()
```

Need to change typos of maxda, porcshce, toyouta, vokswagen, and vw

```python
df["CarName"] = df["CarName"].replace({"maxda":"mazda", "porcshce":"porsche", "toyouta":"toyota",
                                      "vokswagen":"volkswagen", "vw":"volkswagen"})
df["CarName"].unique()
```

**Plot price against all numerical features to see correlation**

```python
#get all numeric features names except last since it's the price
numeric_features = df.select_dtypes(include=["int64", "float64"]).columns.to_list()[:-1]

#plot price vs all the numeric features
fig, axes = plt.subplots(nrows=5, ncols=3, figsize=(20,20))

#turn off last grid since only 14 subplots used on a 5x3 plot
axes[4,2].axis('off') 

#iterate through feature columns and plot against the price
for i, feature in enumerate(numeric_features):
    ax = axes[i // 3, i % 3]  # // to get row index, % to get column index
    ax.scatter(df[feature], df["price"], label=f'{"price"} vs {feature}')
    ax.set_xlabel(feature)
    ax.set_ylabel("price")
    ax.legend()

plt.tight_layout()
plt.show()
```

Symboling, car height, stroke, compression ratio, and peak rpm all have weak correlation with price, so it is reasonable to drop these columns as they will not have much effect on predicting the car price.

```python
to_drop = ["symboling", "carheight", "stroke", "compressionratio", "peakrpm"]
df.drop(columns=to_drop, inplace=True)
```

**Compare price with categorical features**

```python
#group by car brands and plot average price of each car brand
car_brand_avg_price = df.groupby("CarName")["price"].mean()
plt.figure(figsize=(8,6))
plt.title("Price vs car brands")
plt.xlabel("Price")
plt.ylabel("Car brand")
plt.barh(car_brand_avg_price.index, car_brand_avg_price.values)
```

```python
#inspect other categorical features
categorical = df.select_dtypes(include=["object"]).columns.to_list()[1:] #exclude car brand

#plot price vs the categorical features
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20,15))

#iterate through feature columns and plot against the price
for i, feature in enumerate(categorical):   
    ax = axes[i // 3, i % 3] # // to get row index, % to get column index
    sns.boxplot(data=df, x=feature, y="price", ax=ax)
    ax.set_xlabel(feature)
    ax.set_ylabel("price")
    
plt.tight_layout()
plt.show()
```

Door number and fuel systems seem to have a weak relation with price.

```python
to_drop = ["doornumber", "fuelsystem"]
df.drop(columns=to_drop, inplace=True)
```

## Preprocessing

```python
from sklearn.preprocessing import LabelEncoder

#encode categorical variables
categoricals = df.select_dtypes(include=["object"]).columns.to_list()
encoder = LabelEncoder()

for feature in categoricals:
    df[feature] = encoder.fit_transform(df[feature])
```

```python
df.dtypes
```

## Comparing multiple models

```python
from sklearn.model_selection import train_test_split

X = df.drop("price", axis=1)
y = df["price"]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
#compare lasso, random forest, and linear
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.model_selection import KFold, cross_val_score

models = {"Linear Regression":LinearRegression(), "Lasso":Lasso(alpha=0.1), "Random Forest":RandomForestRegressor()}

scores = []
#loop through each model
for model in models.values():
    kf = KFold(n_splits=5, random_state=42, shuffle=True)
    cv_scores = cross_val_score(model, X_train, y_train, cv=kf) #do cross validation
    scores.append(cv_scores)
    
# plot the results
plt.boxplot(scores, labels=models.keys())
plt.show()
```

Linear regression and lasso are equally not good for this problem, while random forest seems to be best.

## Model and evaluation

```python
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)

y_preds = rf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

preds = rf.predict(X_test)
print("mean absolute error: ", mean_absolute_error(y_preds, y_test))
print("mean squared error: ", mean_squared_error(y_preds, y_test))
print("r2_score: ", r2_score(y_preds, y_test))
```

```python
#evaluate the importance of features
feature_importances = rf.feature_importances_

#create dict with feature names corresponding to its importance
feature_importance_dict = {"Feature": df.columns[:-1], "Importance": feature_importances}
feature_importance_df = pd.DataFrame(feature_importance_dict) #create dataframe
feature_importance_df = feature_importance_df.sort_values(by="Importance", ascending=True) #sort by highest importance

#plot importances
plt.figure(figsize=(6,5))
plt.barh(feature_importance_df["Feature"], feature_importance_df["Importance"])
plt.title("Feature Importance")
plt.xlabel("Importance")
plt.ylabel("Feature")
```

Engine size and curb weight seem to hold most importance in car price.
